```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

```

```python
# load image path
import cv2 as cv
from keras.utils.image_utils import load_img,img_to_array
img_folder = Path("/home/tan/Documents/AI_thaythinh/homework/classAI/data/datak20/Data_AI_k20")

label = ["Ai","Tan","Thoai",'Thuong']
test_image = list([])
X_data = np.empty((0,32*32))
Y_label = []

for i in range(len(label)):
    train_folder = img_folder/label[i]
    
    test = list(train_folder.glob("*.jpg"))
    
    test_image = test_image + test
    Y_label = Y_label+[i]*len(list(train_folder.glob("*.jpg")))

for  i in range(len(test_image)):
    image_path = str(test_image[i])
    img = load_img(image_path,target_size=(32,32),color_mode = "grayscale")
    img= img_to_array(img)
    img = img.astype("float32")/255
   

    img = np.reshape(img,(1,32*32))
    X_data = np.vstack((X_data,img))
print(X_data.dtype)
```

```python
folder = Path("classAI/data/datak20/Data_AI_k20/Thoai")
print(list(folder.glob("*.jpg")))
```

```python
from keras.models import Sequential
from keras import layers

model = Sequential()
model.add(layers.Flatten(input_shape=[32*32]))
model.add(layers.Dense(512,activation="relu"))
model.add(layers.Dense(300,activation="relu"))
model.add(layers.Dense(100,activation="relu"))
model.add(layers.Dense(4,activation="softmax"))

model.summary()
```

```python
print(Y_label)
```

```python
model.compile(optimizer= "rmsprop",loss = "categorical_crossentropy",metrics=["accuracy"])
```

```python
from keras.utils import to_categorical
Y_label = to_categorical(Y_label)
print(Y_label)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_data,Y_label,test_size=0.3,shuffle=True)
```

```python
model.fit(X_train,Y_train,epochs =200,batch_size =128,validation_data=[X_test,Y_test])
```
